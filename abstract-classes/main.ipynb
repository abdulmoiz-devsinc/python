```python
from abc import ABC,abstractmethod
```

```python
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Toyota(Vehicle):
    def engine(self):
        print("engine started")

t=Toyota()
```

```python
import threading
import time

def worker(name):
    print(f"Thread {name} starting")
    time.sleep(2)
    print(f"Thread {name} finished")

threads = []

for i in range(3):
    t = threading.Thread(target=worker, args=(i,))
    threads.append(t)
    t.start()  # actually start the thread

for t in threads:
    t.join()   # wait for all threads to finish

print("All threads done")
```
