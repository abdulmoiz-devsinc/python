```python
def auth(func):
    def wrapper(*args, **kwargs):
        # name comes from kwargs
        if kwargs.get("name") == "Abdul Moiz":
            print("User authenticated")
            return func(*args, **kwargs)
        else:
            print("Not authorized")
    return wrapper

@auth
def greet(name):
    print("HELLO " + name)


```

```python
greet(name="Abdul Moiz")
```
