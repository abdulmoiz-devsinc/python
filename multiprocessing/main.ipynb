```python
from multiprocessing import Process, Queue

def worker(q, n):
    q.put(n * n)

if __name__ == "__main__":
    q = Queue()
    processes = [Process(target=worker, args=(q, i)) for i in range(5)]
    
    for p in processes:
        p.start()
    for p in processes:
        p.join()
    
    results = [q.get() for _ in range(5)]
    print(results)
```

```python
from multiprocessing import Process, Value, Lock

def worker(n, lock):
    for _ in range(1000):
        with lock:
            n.value += 1

if __name__ == "__main__":
    n = Value('i', 0)
    lock = Lock()
    processes = [Process(target=worker, args=(n, lock)) for _ in range(5)]
    
    for p in processes:
        p.start()
    for p in processes:
        p.join()
    
    print(n.value)  # 5000
```
