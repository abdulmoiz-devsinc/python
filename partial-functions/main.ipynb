```python
def greet(first_name,last_name,company):
    print("Hello "+first_name+" "+ last_name+" from "+company)
```

```python
from functools import partial
```

```python
devsinc_emp=partial(greet,company="Devsinc")
sys_emp=partial(greet,company="Systems")

```

```python
devsinc_emp("Abdul","Moiz")
```

```python
sys_emp("Hamza","Imran")
```
